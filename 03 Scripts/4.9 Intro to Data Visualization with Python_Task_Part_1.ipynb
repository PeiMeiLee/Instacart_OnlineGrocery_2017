**5. Complete the fundamental data quality and consistency checks youâ€™ve learned throughout this Achievement; for example, check for and address missing values and duplicates, and convert any mixed-type data.**

```python
# Check for NaN values

customers.isnull().sum()
```

```python
# Impute NaN values with "Not Provided"

customers['first_name'].fillna('Not Provided', inplace = True)
```

```python
# Check the output

customers.isnull().sum()
```

```python
# Check for mixed-type data

for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
```

```python
# Change data type of "first_name" column to string

customers['first_name'] = customers['first_name'].astype('str')
```

```python
# Check the output

for col in customers.columns.tolist():
  weird = (customers[[col]].applymap(type) != customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (customers[weird]) > 0:
    print (col)
else:
        print('All good.')
```
